```python
df = df.drop(columns=['flat_model', 'lease_commence_date', 'remaining_lease','block', 'street_name'])

df = pd.get_dummies(df, columns=['town', 'flat_type', 'storey_range'])
```

Certain columns were dropped because they either do not contribute meaningfully to resale price prediction or would introduce unnecessary complexity:

`block` and `street_name` were removed due to:
  - High cardinality with low predictive value.
  - Encoding them would significantly increase dimensionality without meaningful gain in prediction.
  - They introduce noise: resale price is more strongly tied to the **town**, **flat_type**, and **storey range**, rather than exact street names.

lease_commence_date, remaining_lease: Redundant as year/month trends are captured separately.

flat_model: Inconsistent across records, removed for simplicity.

One-hot encoding was used because town, flat_type, and storey_range are categorical without ordinal relationships. This ensures the machine learning models treat these as distinct categories without inferring any numerical hierarchy.

Dates were converted to year and month_num because ML models can't handle dates directly. These new features make it easier to capture trends (e.g., prices increasing over the years). Dropping the raw month column prevents redundancy.

```python
X = df.drop('resale_price', axis=1)
Y = df['resale_price']
df
```

Separating features (X) and target (Y) is essential for supervised learning. resale_price is what we're trying to predict.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)   
```

Splitting the data helps evaluate the model’s generalizability. 70% is used for training, and 30% for testing to ensure performance is not measured on seen data.

GridSearch systematically tests different combinations of hyperparameters. Optimizing n_estimators (trees) and max_depth (tree depth) improves performance and reduces overfitting.

the code 

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Define regressors
models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'DecisionTreeRegressor': DecisionTreeRegressor(),  
    'RandomForestRegressor': RandomForestRegressor(),
    'GradientBoostingRegressor': GradientBoostingRegressor(),
    'KNeighborsRegressor': KNeighborsRegressor()
}

print("\regression Results:")
model_names = []
r2_scores = []

# Evaluate regressors
print("\nRegression Results:")
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    mae = mean_absolute_error(y_test, preds)
    r2 = r2_score(y_test, preds)
    model_names.append(name)
    r2_scores.append(r2)
    print(f"{name}: RMSE = {rmse:.2f}, MAE = {mae:.2f}, R² = {r2:.4f}")
# Visualize the results
```

I used RMSE, MAE, and R² for evaluating model performance:

R² (R-squared) shows how much variance is explained by the model.

RMSE (Root Mean Squared Error) penalizes larger errors more heavily.

MAE (Mean Absolute Error) shows average magnitude of errors without considering direction.

Based on the dataset, i chosen to use regressor models over classfier models because they help better for predicting the resale prices.

Linear Regression: Baseline, simple model, checks linear relationship
Ridge / Lasso: Handles multicollinearity; Lasso performs feature selection
Decision Tree: Captures non-linear relationships easily
Random Forest: Reduces overfitting via averaging; robust performance
Gradient Boosting: Focuses on minimizing errors; powerful ensemble method
KNN Regressor: Simple baseline for localized predictions

- High accuracy
- Low variance
- Better handling of complex feature interactions

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.barh(model_names, r2_scores, color='skyblue')
plt.xlabel('R² Score')
plt.title('Model Performance Comparison')
plt.xlim(0, 1)  # To keep x-axis standard for R² (0 to 1)
plt.show()
```

Bar plots visually compare model performance. Using R² scores makes it clear which model fits the data best.
